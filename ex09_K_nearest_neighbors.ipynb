1 Import all necessary libraries. Then load the dataset and save it to the dataframe.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import neighbors
from sklearn.model_selection import train_test_split

df = pd.read_csv("./data_files/voice.csv")

df = pd.DataFrame(df)

df

```

2 Filter the data so only the following columns will be selected: **meanfun** and **IQR**. Split the dataset so 75 % of the data will be used for training.

```python
data = df[["meanfun", "IQR"]]

# Prepare data
X = df[["meanfun", "IQR"]].to_numpy().astype(float)
y = df["label"].to_numpy()
h= .01

# Split dataset
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)

data
```

3 Find the optimal value for k comparing possible values and error rate.

```python
np.warnings.filterwarnings("ignore", category=FutureWarning)

error_rate = []
for i in range(1,40):
    knn = neighbors.KNeighborsClassifier(n_neighbors = i)
    knn.fit(x_train, y_train)
    pred_i = knn.predict(x_test)
    error_rate.append(np.mean(pred_i != y_test))

plt.figure(figsize = (10,6))
plt.plot(range(1,40), error_rate,color = 'blue', linestyle = 'dashed', marker='o', markerfacecolor='pink', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('k')
plt.ylabel('Error Rate')
min_error = error_rate.index(min(error_rate))
print("A minimum error of", min(error_rate), "is achieved with the k value of", min_error)
```

4 Define a function and include the following operations inside the function:
   - Define a classifier
   - Fit the data for the classifier
   - Calculate minimum and maximum values for boundaries for meanfun and IQR
   - Run predictions
   - Draw a chart from predicted data utilising previously defined colormaps

Utilise your function and call it using the optimal k value chosen previously.

```python
# Define  function
def voices(k):

    # Prepare data
    X = data[["meanfun", "IQR"]].to_numpy().astype(float)
    y = data["label"].replace({"male": 0, "female": 1}).to_numpy()
    h = .01

    # Split dataset
    x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)

    # Define classifier
    knn = neighbors.KNeighborsClassifier(n_neighbors = k)

    # Fit
    knn.fit(x_train, y_train)

    # Calculate minimum and maximum values for boundaries
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

    # Generate grid coordinates for chart
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    # Predict class using data and KNN classifier
    Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
    
    # Colour plot
    bg_colors = ListedColormap(["#FFAAAA", "#AAFFAA"])
    p_colors = ListedColormap(["#FF0000", "#00FF00"])
    
    # Result rendered in colour plot
    Z = Z.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx, yy, Z, cmap=bg_colors)
    
    # Plot training points
    plt.scatter(x_test[:, 0], x_test[:, 1], c=y_test, cmap=p_colors, edgecolors='k', s=20)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title("Voice Classification")
    plt.xlabel("meanfun")
    plt.ylabel("IQR")
    plt.show()

# Call function utilizing optimal k value
optimal_k = 12
voices(optimal_k)
```
